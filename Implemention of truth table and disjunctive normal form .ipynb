```python
def disjunction(a,b):
    
    list_ret = []
    it = len(a)
    i = 0
    while(i < it):
        if(a[i]+b[i] == 0):
            list_ret += [0]
        else:
            list_ret += [1]
        i += 1
    
    return list_ret
```

```python
def conjunction(a,b):
    
    list_ret = []
    it = len(a)
    i = 0
    while(i < it):
        if(a[i]*b[i] == 0):
            list_ret += [0]
        else:
            list_ret += [1]
        i += 1
    
    return list_ret
```

```python
def negation(a):
    
    list_ret = []
    it = len(a)
    i = 0
    while(i < it):
        if(a[i]==1):
            list_ret += [0]
        else:
            list_ret += [1]
        i += 1
    
    return list_ret
```

```python
def implication(a,b):
    
    list_ret = []
    
    neg_a = negation(a)
    list_ret = disjunction(neg_a,b)
    
    return list_ret
```

```python
def bi_implication(a,b):
    
    list_ret = []
    
    list_ret = conjunction(implication(a,b),implication(b,a))
    
    return list_ret
```

```python
def precedence(op):
    
    if op == '=':
        return 1
    if op == '*':
        return 2
    if op == '+':
        return 3
    if op == '.':
        return 4
    if op == '~':
        return 5
    
    return 0;
```

```python
def applyconnectivies(a,b,op):
    
    if op == '=':
        return bi_implication(b,a)
    if op == '*':
        return implication(b,a)
    if op == '+':
        return disjunction(b,a)
    if op == '.':
        return conjunction(b,a)
    if op == '~':
        return negation(a)
```

```python
def evalute(tokens):
    
    #storing list of lists, where list has 0,1s of a variable or an inside bracket
    values = []
    
    #storing operator
    ops = []
    i = 0
    
    while i < len(tokens):
#         print(tokens[i])
        #if current token is a whitespace,then skip it.
        if tokens[i] == ' ':
            i += 1
            continue
        
        
        elif tokens[i] == '(':
            ops.append(tokens[i])
            
        
        #Appends to list!!!
        elif type(tokens[i]) == list :
#             print('var :',tokens[i])
            values.append(tokens[i])
            
            
        #Closing bracket encounter
        #Solve entire braces untill closing bracket is found.
        elif tokens[i] == ')':
            
            while len(ops)!=0 and ops[-1] != '(':
                
                list1 = values.pop()
                
                op = ops.pop()
                if op == '~':
                    list2 = []
                else:
                    list2 = values.pop()
            
                values.append(applyconnectivies(list1,list2,op))
                
            #Poping the opening brace    
            ops.pop()    
        
        
        #if we have an operator , then check the top of the stack and decide to push or not
        else:
            
            while(len(ops) != 0 and precedence(ops[-1]) >= precedence(tokens[i])):
                
                list1 = values.pop()
                
                op = ops.pop()
                
                if op == '~':
                    list2 = []
                    
                else:
                    list2 = values.pop()
                
                values.append(applyconnectivies(list1,list2,op))
                
            #push current token to ops.
            ops.append(tokens[i])
            
            
            
        i += 1     
            
    #for remaining operator, pop the lists and apply these operator
    
    while len(ops)!= 0:
        
        list1 = values.pop()
        op = ops.pop()
                
        if op == '~':
            list2 = []
        else:
            list2 = values.pop()
        
        values.append(applyconnectivies(list1,list2,op))
        
        
            
    return values[0]           
```
